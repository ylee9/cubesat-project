## Setting Up Problem

Area under a sine wave, between theta = 0 and pi

```python
theta = np.linspace(0,np.pi, 50)
y = np.sin(theta)

fig, ax = plt.subplots()
ax.plot(theta,y)
ax.fill_between(theta, y, alpha=0.5)
ax.set(xlabel=r'$\theta$', ylabel=r'$\sin(\theta)$', ylim=(0,1.05))
```

```python
def f(theta):
    return np.sin(theta)

numerical_result = integrate.quad(f,0,np.pi)

print numerical_result[0]
#second element in the list would be the error estimate which is redundant in this case
```
